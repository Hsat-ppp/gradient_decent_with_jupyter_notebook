I import libraries.


```python
from matplotlib import pyplot as plt
```

I define the test function to be optimized, and its derivative.


```python
def test_function(x):
    return x**2


def test_function_derivative(x):
    return 2 * x
```

... and update rule based on gradient decent.


```python
def update(x, alpha=0.1):
    d = test_function_derivative(x)
    x -= alpha * d
    return x
```

Now, optimization procedure is applied.


```python
x = 10  # arbitrary initial value
num_of_iteration = 100  # number of iteration
stop_criterion = 1e-3  # stop criterion
history_dict = {
    'x': [x],
    'function_value': [test_function(x)]
}
for _ in range(num_of_iteration):
    x = update(x)  # update x
    function_value = test_function(x)
    history_dict['x'].append(x)
    history_dict['function_value'].append(function_value)
    if function_value < stop_criterion:  # if True, stop procedure.
        break
```

Finally, the optimization history is visualized.


```python
for x_now, f_now in zip(history_dict['x'], history_dict['function_value']):  # better writing?
    plt.plot(history_dict['x'], history_dict['function_value'])
    plt.plot(x_now, f_now, 'o')
    plt.show()
```
